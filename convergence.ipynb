```python
# Dataset and Initial Parameters
x = [
    [1, 8],
    [6, 7],
    [8, 5],
    [11, 4],
    [6, 2],
    [3, 3],
    [2, 4],
    [11, 2],

]

# lr = 0.2
b = -10
w = [1, 1]

```

```python
#Augment Bias into x and w 



for row in x : 
    row.append(1)

x

w.append(b)


# Plotting the data
import matplotlib.pyplot as plt


for row in x :
    plt.plot(row[0], row[1], 'ro')
    
plt.show()
```

```python





for row in range(4, 8):
    for col in range(len(x[row])):
        x[row][col] = -x[row][col]

print(x)
print(w)

```

```python
epochs = 10
lr = [0.2, 0.1, 0.01,  0.001]
```

```python
def calculate_column_sums(matrix):
    num_rows = len(matrix)
    if num_rows == 0:
        return []  # Return an empty list if the matrix is empty
    num_cols = len(matrix[0])

    # Initialize a list to store column sums, initially filled with zeros
    column_sums = [0] * num_cols


    for row in matrix:
        for col in range(num_cols):
            column_sums[col] += row[col]

    return column_sums



```

```python



for l in lr:

    
    print("_____________________________________Learning Rate : ", l,'_____________________________________')
    for epoch in range(1, epochs + 1):
        print("---------------------------------------------")
        print("Epoch : ", epoch)

        y = []
        wrong_x = []
        delta = [0] * len(w)
        for row in x:
            y1 = round(row[0] * w[0] + row[1] * w[1] + row[2] * w[2], 3)  

            if y1 < 0:
                wrong_x.append(row)
            y.append(y1)
        print("F(x) values are  : ", y)

        delta = calculate_column_sums(wrong_x)
        
        
        print("Delta values are : ", delta)

        for i in range(len(w)):
            w[i] = round(w[i] + l * delta[i], 3)  

        print('Updated weights = ', w)



```
